```python
# check popular hrs by plotting number of rides per hr on time axis
# using grouping pandas equipment
df['lpep_pickup_datetime'].head().dt.round('H') # round time in nearest hrs
```

```python
# use hours as groupby key
keys = df['lpep_pickup_datetime'].dt.round('H')
df.groupby(keys)
```

```python
df.groupby(keys).count().head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (20, 8)
```

```python
df.groupby(keys).count()['Vendor'].plot()
```

```python
# plot in a single day(rides per day in hr group)
df.groupby(keys).count()['Vendor'].loc['2015-03-10'].plot.bar()
```

```python
df['hour'] = df['lpep_pickup_datetime'].dt.hour
df['day'] = df['lpep_pickup_datetime'].dt.date
df[['hour', 'day']].head()
```

```python
df.groupby(['Vendor', 'day', 'hour']).count().head()
```

```python
df.groupby(['Vendor', 'day', 'hour']).count().index
```

```python
ddf = df.groupby(['Vendor', 'day', 'hour'], as_index=False).count()
ddf.head()
```

```python
# grouping medium rides in a day in every hours 
hdf = ddf.groupby(['Vendor', 'hour'], as_index=False).median()
hdf.head()
```

```python
vdf = hdf.pivot(columns='Vendor', index='hour', values='Extra')
vdf.plot.bar(rot=45)
```

```python
import sqlite3
conn = sqlite3.connect('weather.db')
```

```python
wdf = pd.read_sql('SELECT * FROM weather', conn)
wdf.columns
```

```python
wdf = pd.read_sql('SELECT * FROM weather', conn, parse_dates=['DATE'], index_col=['DATE'])
wdf.index
```

```python
wdf.describe()
```

```python
# convers Temp to Farenheit using scipy
from scipy.constants import convert_temperature
wdf['tempF'] = convert_temperature(wdf['TMAX']/10, 'C', 'F')
wdf.head()
```

```python
ddf = df.groupby(df['lpep_pickup_datetime'].dt.date).count() #group rides in pickup datetime in dates
```

```python
# join weather wdf and ddf rides
jdf = ddf.join(wdf)
jdf.head()
```

```python
# plot by scatter plot x=tempF, y=Vendor
jdf.plot.scatter(x='tempF', y='Vendor')
```

```python
jdf.plot.scatter(x='tempF', y='Vendor').set_ylabel('Daily Rides')
```
